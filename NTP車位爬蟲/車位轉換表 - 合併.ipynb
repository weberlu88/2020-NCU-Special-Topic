```python
import pandas as pd

df = pd.read_csv('轉換表.csv')
df
```

```python
df_new = pd.read_csv('轉換表_new.csv')
df_new
```

* Merge, join, and concatenate https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
* pandas.DataFrame.rename https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rename.html

```python
df_concat = pd.concat([df, df_new], ignore_index=True, axis=1)
```

```python
df_concat
```
