這支程式會計算道路上有幾個車位，再給道路一個座標。
座標是取每個停車格座標的中位數。

```python
import pandas as pd
```

```python
df_org = pd.read_excel('NTP_all.xlsx', sheet_name=0)
df_org.head(3)
```

```python
df_route = pd.read_excel('NTP_all.xlsx', sheet_name=2)
df_route.head(3)
```

```python
street_names = pd.unique(df_org['route'])
len(street_names)
```

```python
lat, lon = df_route.iloc[1][['lat','lon']]
lon
```

### 計算每條路有幾個車位

```python
df = pd.DataFrame()
df['roadid'] = ""
df['lat'] = ''
df['lon'] = ''
df['route'] = ''
df['車格數量'] = ''
```

```python
for street in street_names:
    df_select = df_org.loc[df_org['route'] == street]
#     print(street, len(df_select))
    df = df.append({'route':street, '車格數量':len(df_select)}, ignore_index=True)
```

```python
df
```

### 幫路找一個座標 (取中位數)

```python
for street in street_names:
    df_select = df_org.loc[df_org['route'] == street]
    df_select.sort_values(by=['lat', 'lon'], inplace=True)
    
    _len  = len(df_select)
    _mean = int(_len/2)
    
    lat, lon = df_select.iloc[_mean-1][['lat','lon']]
    df.loc[df.route == street, 'lat'] = lat
    df.loc[df.route == street, 'lon'] = lon
    
```

```python
df.roadid = list(range(1, len(df)+1))
df
```

```python
df.to_csv('tbl路段_板橋.csv', index=False)
```
