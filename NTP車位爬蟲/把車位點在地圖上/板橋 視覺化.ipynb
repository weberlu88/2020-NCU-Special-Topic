```python
import pandas as pd
```

```python
df = pd.read_excel('NTP_All.xlsx')
df_route = df.loc[df['route'] == '光中路']
len(df_route)
```

```python
df_route
```

```python
lon_min = df['lon'].min()
lon_max = df['lon'].max()
lat_min = df['lat'].min()
lat_max = df['lat'].max()

print('lon_min:', lon_min)
print('lon_max:', lon_max)
print('lat_min:', lat_min)
print('lat_max:', lat_max)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=df_route['lon'], y=df_route['lat'], s=3)
plt.show()
```

### define the Bounding Box

```python
# BBox = ((df.longitude.min(),   df.longitude.max(),      
#          df.latitude.min(), df.latitude.max())
BBox = ((121.4243, 121.4889,      
         24.9718, 25.0402))

ruh_m = plt.imread('板橋圖2.png')
```

### Plotting

```python
fig, ax = plt.subplots(figsize = (8,6)) # 4:3
ax.scatter(df_route['lon'], df_route['lat'], zorder=1, alpha= 0.2, c='b', s=2)
ax.set_title('Plotting Spatial Data on Riyadh Map')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(ruh_m, zorder=0, extent = BBox, aspect= 'equal')
fig.savefig('filename.png', dpi=300)
```
